```python
import gensim
from gensim.models import word2vec
import nltk

#sample documents
documents= ["this is the first document",
    "this document is the second document",
    "and this is the third one",
    "is this the fourth document."
]

#preprocess the documents: tokenize the lowercase
nltk.download('punkt')
tokenized_docs= [nltk.word_tokenize(d.lower()) for d in documents]
#train the word2vec model
# Use the lowercase 'word2vec' to match the import
model= word2vec.Word2Vec(sentences= tokenized_docs,vector_size=300,window=10,min_count=1,sg=0)
#access the vector for a specific word
cat_vector= model.wv['document']
print("vector representation for 'document':\n",cat_vector)
#finding the most similar words to 'cat'
similar_words= model.wv.most_similar('document')
print("most similar words to 'document':\n",similar_words)
```
